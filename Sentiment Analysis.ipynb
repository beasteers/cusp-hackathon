```python
import got
import nlp
import nltk
import pandas as pd
from nltk.tokenize import TweetTokenizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Get tweets
tweetCriteria = got.manager.TweetCriteria().set(
	q='"Harvey Weinstein"', max_tweets=20
)

pulled_tweets = got.manager.TweetManager.getTweets(tweetCriteria)
len(pulled_tweets), pulled_tweets[0].__dict__
```

```python
pd.DataFrame([t.__dict__ for t in pulled_tweets]).head()
```

```python
tknzr = TweetTokenizer(strip_handles=True, reduce_len=True)
sid = SentimentIntensityAnalyzer()
for tweet in pulled_tweets:
    print(tweet.text)
    print(nlp.tweet_tokenize(tweet.text))
    ss = sid.polarity_scores(tweet.text)
    print(', '.join(['{}: {}'.format(k, ss[k]) for k in ss]))
    print
```
